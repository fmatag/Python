```
#Operadores
a = 10
b = 3

# aritméticos
print("Operadores aritméticos")
print("==============================================================")
print("Suma: %d + %d =" % (a, b), a + b)
print("Resta: %d - %d =" % (a, b), a - b)
print("Multiplicación: %d * %d =" % (a, b), a * b)
print("División: %d / %d =" % (a, b), a / b)
print("División entera: %d // %d =" % (a, b), a // b)
print("Módulo: %d %% %d =" % (a, b), a % b)
print("Potencia: %d ** %d =" % (a, b), a ** b)

a += 10
print(a)

# relacionales
print("\nOperadores relacionales")
print("==============================================================")
print("Igual: %d == %d es" % (a, b), a == b)
print("Diferente: %d != %d es" % (a, b), a != b)
print("Mayor que: %d > %d es" % (a, b), a > b)
print("Menor que: %d < %d es" % (a, b), a < b)
print("Mayor o igual: %d >= %d es" % (a, b), a >= b)
print("Menor o igual: %d <= %d es" % (a, b), a <= b)

# lógicos
print("\nOperadores lógicos")
print("==============================================================")
print("NOT: not(%d == %d) es" % (a, b), not(a == b))
print("AND: (%d > %d) and (%d == %d) es" % (a, b, a, b), (a > b) and (a == b))
print("OR: (%d > %d) or (%d == %d) es" % (a, b, a, b), (a > b) or (a == b))

# ejemplo de variables por referencia
print(a, b)
a = b
print(a, b)
a += 10
print(a, b)
```

```
# Hileras
s = "Hola estimados participantes"
s1 = "Bienvenidos al curso virtual"

print(s + s1)
print("%s %s" % (s, s1.lower()))

# repeticiones
print("\nHILERAS: repeticiones")
print("==============================================================")
print("'A' * 5" , "A" * 5, sep=" => ")
print("20 * 'B'" , 20 * "B", sep=" => ")

# información
print("\nHILERAS: obtener información")
print("==============================================================")
print("'a'.isdigit() :", "a".isdigit())
print("'a'.isalpha() :", "a".isalpha())
print("'a2'.isalnum() :", "a2".isalnum())
print("'hola'.islower() :", "hola".islower())
print("'HOLA'.isupper() :", "HOLA".isupper())
print("'Hola mundo'.istitle() :", "Hola mundo".istitle())
print("'    '.isspace() :", "   ".isspace())
print("'A   '.isspace() :", "A  ".isspace())

# count y len
print("\nHILERAS: ocurrencias y longitud")
print("==============================================================")
print("'Hola mundo'.count('o') :", "Hola mundo".count("o"))
print("len('Hola mundo') :", len("Hola mundo"))

# modificar hilera
print("\nHILERAS: modificar")
print("==============================================================")
print("title()", "Hola estimados participantes".title(), sep=" => ")
print("capitalize()", "Hola estimados participantes".capitalize(), sep=" => ")
print("lower()", "Hola estimados participantes".lower(), sep=" => ")
print("upper()", "Hola estimados participantes".upper(), sep=" => ")
print("swapcase()", "Hola estimados participantes".swapcase(), sep=" => ")

# modificar hilera
print("\nHILERAS: agregar caracteres")
print("==============================================================")
print("center()", "hola".center(20, "*"), )
print("ljust() ", "hola".ljust(20, "*"), sep=" => ")
print("rjust() ", "hola".rjust(20, "*"), sep=" => ")
print("zfill() ", "hola".zfill(20), sep=" => ")

# modificar hilera
print("\nHILERAS: eliminar espacios")
print("==============================================================")
print("' hola '.strip()", " hola ".strip(), sep=" => ")
print("' hola '.lstrip() ", " hola ".lstrip(), sep=" => ")
print("' hola '.rstrip() ", " hola ".rstrip(), sep=" => ")
print("' hola'.rstrip('la') ", " hola".rstrip("la"), sep=" => ")

# reemplazo de hileras
print("\nHILERAS: reemplazo")
print("==============================================================")
print("' hola '.replace(' ', '*')", " hola ".replace(" ", "*"), sep=" => ")
print("' hola '.replace(' ', '*', 1)", " hola ".replace(" ", "*", 1), sep=" => ")


intab = "aeiou"
outtab = "12345"
trantab = str.maketrans(intab, outtab)
print("' hola '.translate('aeiou, '12345')", " hola ".translate(trantab), sep=" => ")


# búsqueda
print("\nHILERAS: búsqueda")
print("==============================================================")
print("'hola'.upper().startswith('H')", "hola".upper().startswith('H'), sep=" => ")
print("'hola'.endswith('z')", "hola".endswith('z'), sep=" => ")
print("'hola'.find('l')", "hola".find('l'), sep=" => ")
print("'hola'.find('z')", "hola".find('z'), sep=" => ")
print("'hola'.index('l')", "hola".index('l'), sep=" => ")
#print("'hola'.index('z')", "hola".index('z'), sep=" => ")
print("string[start:end]", "'hola'[1:3]", "hola"[1:3], sep=" => ")
print("string[start:]", "'hola'[2:]", "hola"[2:], sep=" => ")
print("string[:end]", "'hola'[:2]", "hola"[:2], sep=" => ")
print("string[:end]", "'hola'[:-1]", "hola"[:-1], sep=" => ")
print("string[start:end:step]", "'hola'[::2]", "hola"[::2], sep=" => ")

mensaje = "Hola Mundo"
if 'o' in mensaje:
    print("xxxx")
```

```
# Colecciones
# funcionalidades comunes en colecciones
print("\nCOLECCIONES: funcionalidades comunes")
print("==============================================================\n")
print("**** RECORRER UNA COLECCION")
print("for valor in [1,2,3]:\n    print(valor)\n\nResultados:")
for valor in [1,2,3]:
    print(valor)

print("\n**** EXISTE UN ELEMENTO EN LA COLECCION")
print("if 3 in (1,2,3):\n    print('3 si está en la tupla')\n\nResultados:")
if 3 in (1,2,3):
    print('3 si está en la tupla')

print("\n**** CANTIDAD DE ELEMENTOS EN LA COLECCION")
print("len((1,2,3))\n\nResultados:")
print( len( (1,2,3) ) )
```
